```python
import numpy as np
```

```python
from matplotlib import pyplot as pt
```

```python
x=np.arange(1,11)
```

```python
x
```

```python
y=2*x
```

```python
pt.plot(x,y)
pt.show()
```

```python
pt.plot(x,y,color='g',linestyle=':',linewidth=5)
pt.title('x vs y')
pt.xlabel('X axis')
pt.ylabel('Y Axis')
pt.grid(True)
pt.show()
```

```python
student={'bob':55,'ram':56,'anita':100,'sarita':68}
```

```python
keys=list(student.keys())
```

```python
keys
```

```python
marks=list(student.values())
```

```python
marks
```

```python
pt.bar(keys,marks,color='g')
pt.title('student marks')
pt.xlabel('name of student')
pt.ylabel('obtain marks')
pt.show()
```

```python
fruit=['apple','mango','banana','orange']
```

```python
quantity=[1,4,6,10]
```

```python
pt.pie(quantity,labels=fruit)
pt.show()
```

```python
pt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','green','yellow','blue','orange','black'])
pt.show()
```

```python
## doughnut chart
```

```python
pt.pie(quantity,labels=fruit,autopct='%0.1f%%',colors=['red','green','yellow','blue','orange','black'])
pt.pie([1],colors='w',radius=.5)
pt.show()
```
